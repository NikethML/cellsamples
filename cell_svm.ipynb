```python
#Package imports needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import seaborn as sns
from scipy import stats
import scipy
```

```python
url='https://raw.githubusercontent.com/NikethML/cellsamples/main/cell_samples%20(7).csv'
df= pd.read_csv(url)
```

```python
df.head()
```

```python
df.dtypes
```

```python
col_names=df.columns.tolist()
for i in col_names:
  print(df[i].value_counts())
```

```python
# all independent variables are categorical with 10 values
```

```python
df.drop('ID',axis =1 ,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder


# Converting type of columns to category

df=df.astype('category')
  
```

```python
df.dtypes
```

```python
df.columns
```

```python
x=df[['Clump', 'UnifSize', 'UnifShape', 'MargAdh', 'SingEpiSize', 'BareNuc',
       'BlandChrom', 'NormNucl', 'Mit']]
y=df['Class']
```

```python
#Create an instance of One-hot-encoder
enc=OneHotEncoder()
  
#Passing encoded columns

enc_data=pd.DataFrame(enc.fit_transform(x).toarray())

enc_data.columns=enc.get_feature_names_out()
  
#Merge with main
  

enc.get_feature_names_out()
```

```python
y=y.replace([2,4],[0,1])
```

```python
#Splitting with 75% training, 25% testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(enc_data, y, train_size = 0.75,
                                                    test_size = 0.25, random_state = 100)
```

```python
from sklearn.svm import SVC  
clf = SVC(kernel='linear') 
  
# fitting x samples and y classes 
clf.fit(x_train, y_train) 
```

```python
y_pred =clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix


cnf_matrix = confusion_matrix(y_test, y_pred)
from sklearn.metrics import *

print("Accuracy:",accuracy_score(y_test, y_pred)) 

print("Precision:",precision_score(y_test, y_pred)) 

print("Recall:",recall_score(y_test, y_pred)) 
```

```python
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
 #print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```
